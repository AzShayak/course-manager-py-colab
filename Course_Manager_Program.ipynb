```
data = []
choice = 'random'

def show_menu():
    print('\n\nMENU\n')
    print('1. Add course')
    print('2. Unenroll from course')
    print('3. View course list')
    print('4. Exit')
    print('---------------------------')
    choice = input('Enter your choice: ')
    return choice

while choice != '4':
    choice = show_menu()
    if choice == '1':
        item = input('What is the course code? ')
        data.append(item)
        print('Course', item, 'added')
    elif choice == '2':
        item = input('Which course to unenroll? ')
        if item in data:
            data.remove(item)
            print('Removed course', item)
        else:
            print('Could not find course', item)
    elif choice == '3':
        print('List of enrolled courses:')
        for item in data:
            print(item)
    elif choice == '4':
        print('\nProgram ended.')
    else:
        print('Please enter one of 1, 2, 3 or 4')
```
